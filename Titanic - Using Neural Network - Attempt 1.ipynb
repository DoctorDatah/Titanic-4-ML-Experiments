```python
import pandas as pd
import io
import requests
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Loading the Data
df_train = pd.read_csv("Train.csv",na_values=['NA','?'])

#randomly shuffle rows
df_train = df_train.sample(frac=1)
df_train.head()
#df_train.Cabin.unique()
```

```python
#Handles missing vales
df_train['Age'] = df_train['Age'].fillna(df_train['Age'].median())
```

```python
# Create dummy variables for categorical columsn
df_sex = pd.get_dummies(df_train['Sex'])
df_embarked = pd.get_dummies(df_train['Embarked'])
#df_cabin = pd.get_dummies(data_fram['Cabin'])

# Add to data fram
#Concat new columns to original dataframe 
df_train = pd.concat([df_train, df_sex,df_embarked], axis=1) #, df_cabin
df_train.head() 
```

```python
x = df_train[['Pclass','Age','Parch','SibSp','female','male','C','Q','S']].values
x[:5]
```

```python
y = df_train[['Survived']].values
y[:5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=97)
x_train[:5]
```

```python
### Build neural network

model = Sequential()
#Hidden Layer 1
model.add(Dense(20, activation='relu')) # nutorns and activation function
#Hidden Layer 2
model.add(Dense(20, activation='relu'))
# Output
model.add(Dense(1))
#model.compile(loss='binary_crossentropy', optimizer='adam')
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
## Fitting the model
## with early stopping

#model.fit(x,y,verbose=2,epochs=100)
monitor = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=4, verbose=1, mode='min',restore_best_weights=True)
model.fit(x,y,validation_data=(x_test,y_test),callbacks=[monitor],verbose=2,epochs=200)
# stop the model fitting from the point from where the error no longer improves.
```

```python
pred = model.predict(x)
# turn off scientific notation
np.set_printoptions(suppress=True)
pred[:5]
```

```python
## Checking Acurrcy

from sklearn.metrics import accuracy_score
# # whatis the percentage of predictions were correct? 
predict_classes = np.argmax(pred,axis=1)
actual_classes = np.argmax(y,axis=1)
accuracy = accuracy_score(actual_classes,predict_classes)
print(f"Accuracy: {accuracy}")
```
